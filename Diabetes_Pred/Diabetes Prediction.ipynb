```python
import numpy as np
import pandas as pd

train_data = pd.read_csv('diabetes_train_set.csv')
train_data.head()
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
#df.replace(to_replace =["Boston Celtics", "Texas"], value ="Omega Warrior")
train_data.replace(to_replace=["Yes", "No"], value=["1","0"], inplace=True)
train_data.head()
```

```python
train_data.replace(to_replace=["Male", "Female"], value=["1","0"], inplace=True)
train_data.replace(to_replace=["Positive", "Negative"], value=["1","0"], inplace=True)
```

```python
train_data.head()
train_data = train_data.astype('int64')
train_data.head()
train_data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_scaler = MinMaxScaler()
min_data = min_scaler.fit_transform(train_data)
df_min_data = pd.DataFrame(data=min_data, columns=train_data.columns)
df_min_data.head()
```

```python
X_train_min = df_min_data.iloc[:,:16]
Y_train_min = df_min_data.iloc[:, 16:]
Y_train_min.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train_min, X_test_min, Y_train_min, Y_test_min = train_test_split(X_train_min, Y_train_min, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train_min, Y_train_min)
log_predict = logistic_model.predict(X_test_min)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(logistic_model, X_train_min, Y_train_min, scoring="neg_mean_squared_error", cv=10)
final_scores_logistic_min = np.sqrt(-scores)

def display_scores(scores):
    print("Scores : ", scores)
    print("Mean : ", scores.mean())
    print("Standard Deviation : ", scores.std())
    
display_scores(final_scores_logistic_min)

acc = logistic_model.score(X_test_min, Y_test_min)
print("Accuracy of Logistic Regression : ",acc)
```

```python
from sklearn.svm import SVC

svc_model = SVC(kernel='rbf')
svc_model.fit(X_train_min, Y_train_min)
svc_predictions = svc_model.predict(X_test_min)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(svc_model, X_train_min, Y_train_min, scoring="neg_mean_squared_error", cv=10)
final_scores_svc = np.sqrt(-scores)

def display_scores(scores):
    print("Scores : ", scores)
    print("Mean : ", scores.mean())
    print("Standard Deviation : ", scores.std())
    
display_scores(final_scores_svc)

acc = svc_model.score(X_test_min, Y_test_min)
print("Accuracy of SVC(rbf) Regression : ",acc)
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_model = DecisionTreeClassifier(max_depth=10)
b = decision_model.fit(X_train_min, Y_train_min)
decision_predictions = decision_model.predict(X_test_min)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(decision_model, X_train_min, Y_train_min, scoring="neg_mean_squared_error", cv=10)
final_scores_decision = np.sqrt(-scores)

def display_scores(scores):
    print("Scores : ", scores)
    print("Mean : ", scores.mean())
    print("Standard Deviation : ", scores.std())
    
display_scores(final_scores_decision)

acc = decision_model.score(X_test_min, Y_test_min)
print("Accuracy of Decision Tree Classifier : ",acc)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(n_estimators=10, max_leaf_nodes=10)
random_forest_model.fit(X_train_min, Y_train_min)
random_forest_predictions = random_forest_model.predict(X_test_min)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(random_forest_model, X_train_min, Y_train_min, scoring="neg_mean_squared_error", cv=10)
final_scores_random = np.sqrt(-scores)

def display_scores(scores):
    print("Scores : ", scores)
    print("Mean : ", scores.mean())
    print("Standard Deviation : ", scores.std())
    
display_scores(final_scores_random)

acc = random_forest_model.score(X_test_min, Y_test_min)
print("Accuracy of Random Forest Classifier : ",acc)
```

```python
import xgboost

xgb_model = xgboost.XGBClassifier()
xgb_model.fit(X_train_min, Y_train_min)
xgb_predictions = xgb_model.predict(X_test_min)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(xgb_model, X_train_min, Y_train_min, scoring="neg_mean_squared_error", cv=10)
final_scores_xgb = np.sqrt(-scores)

def display_scores(scores):
    print("Scores : ", scores)
    print("Mean : ", scores.mean())
    print("Standard Deviation : ", scores.std())
    
display_scores(final_scores_xgb)

acc = xgb_model.score(X_test_min, Y_test_min)
print("Accuracy of XG-Boosting : ",acc)
```
